###01.What is hypothesis testing in statistics?
Hypothesis testing is a statistical method used to make decisions about a population based on a sample of data. It's a formal procedure for investigating our ideas about the world using statistical evidence. The goal is to determine if there is enough evidence to reject a null hypothesis in favor of an alternative hypothesis.

###03.What is the significance level in hypothesis testing, and why is it important?
The significance level (often denoted as α) is the probability of rejecting the null hypothesis when it is actually true. It is a threshold set by the researcher before conducting the test. Common significance levels are 0.05 (5%), 0.01 (1%), or 0.10 (10%).

Importance:

- It determines the critical region or rejection region for the test statistic.
- It quantifies the risk of making a Type I error (false positive).
- It helps in making a decision: If the P-value is less than or equal to α, we reject the null hypothesis.

###04.What does a P-value represent in hypothesis testing?
The P-value (probability value) is the probability of observing a test statistic as extreme as, or more extreme than, the one calculated from your sample data, assuming that the null hypothesis is true. It essentially measures the strength of the evidence against the null hypothesis.

###05.How do you interpret the P-value in hypothesis testing?
- If P-value ≤α (significance level): This indicates that the observed data is unlikely to have occurred if the null hypothesis were true. Therefore, we reject the null hypothesis. This suggests that there is statistically significant evidence to support the alternative hypothesis.
- If P-value >α: This indicates that the observed data is likely to have occurred if the null hypothesis were true. Therefore, we fail to reject the null hypothesis. This means there is not enough statistically significant evidence to support the alternative hypothesis.

###06.What are Type 1 and Type 2 errors in hypothesis testing?
- Type I Error (α): Also known as a false positive. This occurs when you reject the null hypothesis when it is actually true. The probability of making a Type I error is equal to the significance level (α).

  Analogy: Convicting an innocent person.
- Type II Error (β): Also known as a false negative. This occurs when you fail to reject the null hypothesis when it is actually false.

  Analogy: Letting a guilty person go free.

###07. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
A one-tailed test (directional) is used when the alternative hypothesis predicts a specific direction for the effect, like "greater than" or "less than." The critical region for rejecting the null hypothesis is located in only one tail of the distribution. A two-tailed test (non-directional), conversely, is used when the alternative hypothesis simply states that a difference or effect exists, without specifying its direction. In this case, the critical region is split between both tails of the distribution.



###08.What is the Z-test, and when is it used in hypothesis testing?
The Z-test is a statistical hypothesis test used to determine if there is a significant difference between a sample mean and a population mean, or between two sample means. It is typically used when the population standard deviation is known, and the sample size is large (generally n>30). The Z-test relies on the assumption that the sampling distribution of the test statistic follows a standard normal (Z) distribution.

###09. How do you calculate the Z-score, and what does it represent in hypothesis testing?
The Z-score is calculated as Z=(X−μ)/σ, where X is the sample mean (or individual data point), μ is the population mean, and σ is the population standard deviation (or standard error of the mean). In hypothesis testing, the Z-score represents how many standard deviations an observed sample mean (or data point) is away from the hypothesized population mean under the null hypothesis. A larger absolute Z-score indicates stronger evidence against the null hypothesis.

###10. What is the T-distribution, and when should it be used instead of the normal distribution?
The T-distribution, also known as Student's T-distribution, is a probability distribution similar to the normal distribution but with heavier tails, meaning it has more probability in the tails and less at the center. It should be used instead of the normal distribution when the population standard deviation is unknown and must be estimated from the sample data. It is particularly relevant for small sample sizes (n<30), as it accounts for the increased uncertainty due to estimating the population standard deviation.

###11. What is the difference between a Z-test and a T-test?
The primary difference lies in the knowledge of the population standard deviation and sample size. A Z-test is used when the population standard deviation is known and/or the sample size is large (n≥30), allowing the use of the normal distribution. A T-test, on the other hand, is employed when the population standard deviation is unknown and has to be estimated from the sample, especially with small sample sizes (n<30), thus relying on the T-distribution, which accounts for greater uncertainty.

###12. What is the T-test, and how is it used in hypothesis testing?
The T-test is a statistical hypothesis test used to compare means. It is particularly useful when comparing a sample mean to a known population mean, or comparing the means of two independent or dependent samples, especially when the population standard deviation is unknown and estimated from the sample. In hypothesis testing, the T-test calculates a t-statistic, which is then compared to a critical value from the T-distribution to decide whether to reject the null hypothesis.

###13. What is the relationship between Z-test and T-test in hypothesis testing?
The T-test can be seen as an adaptation of the Z-test for situations where the population standard deviation is unknown. As the sample size increases, the T-distribution approaches the standard normal (Z) distribution. Therefore, for very large sample sizes, the results of a T-test and a Z-test will be very similar. In essence, the T-test provides a more robust approach when population parameters are not fully known, making it more frequently applicable in real-world scenarios.



###14. What is a confidence interval, and how is it used to interpret statistical results?
A confidence interval is a range of values, derived from sample data, that is likely to contain the true population parameter (e.g., mean, proportion) with a certain level of confidence. For example, a 95% confidence interval means that if we repeated the sampling process many times, 95% of the intervals constructed would contain the true population parameter. It's used to interpret results by providing a plausible range for the true value, rather than just a single point estimate, and to assess the precision of an estimate.

###15. What is the margin of error, and how does it affect the confidence interval?
The margin of error is a quantifiable amount that accounts for random sampling error in a survey or study. It represents the maximum expected difference between the true population parameter and the sample estimate. The margin of error directly affects the width of the confidence interval: a larger margin of error results in a wider confidence interval, indicating less precision in the estimate, while a smaller margin of error yields a narrower, more precise interval. It's typically calculated as the critical value (from Z or T distribution) multiplied by the standard error.



###16. How is Bayes' Theorem used in statistics, and what is its significance?
Bayes' Theorem describes the probability of an event, based on prior knowledge of conditions that might be related to the event. It's mathematically expressed as P(A∣B)=[P(B∣A)∗P(A)]/P(B). Its significance in statistics lies in its ability to update probabilities based on new evidence. This is foundational for Bayesian statistics, which offers an alternative paradigm to frequentist methods by incorporating prior beliefs into statistical inference, leading to more flexible and intuitive interpretations of probability.

###17. What is the Chi-square distribution, and when is it used?
The Chi-square (χ
2
 ) distribution is a skewed probability distribution that arises in statistical inference, particularly when dealing with sums of squared standard normal variates. It is primarily used in hypothesis testing for analyzing categorical data. Common applications include testing for independence between two categorical variables (Chi-square test of independence) and testing whether observed frequencies differ significantly from expected frequencies (Chi-square goodness-of-fit test). It's always non-negative and defined by its degrees of freedom.

###18. What is the Chi-square goodness of fit test, and how is it applied?
The Chi-square goodness-of-fit test is a non-parametric hypothesis test used to determine how well observed categorical data fits an expected distribution. It assesses whether there is a statistically significant difference between the observed frequencies of categories in a sample and the frequencies that would be expected under a hypothesized distribution. It is applied by calculating a Chi-square statistic based on the observed and expected counts, which is then compared to a critical value from the Chi-square distribution with appropriate degrees of freedom to make a decision.

###19. What is the F-distribution, and when is it used in hypothesis testing?
The F-distribution is a continuous probability distribution that arises in hypothesis testing when comparing variances or when testing the overall significance of a statistical model, particularly in ANOVA. It is defined by two degrees of freedom parameters. It is used when comparing the ratio of two variances, or more commonly, in F-tests within ANOVA to determine if there are statistically significant differences between the means of three or more groups, or to assess the overall fit of a regression model.

###20. What is an ANOVA test, and what are its assumptions?
ANOVA (Analysis of Variance) is a collection of statistical models used to analyze the differences among group means in a sample. It partitions the total variability in a dataset into different sources of variation. Its primary assumptions are: 1) Independence of observations: Samples are drawn independently from each other. 2) Normality: The data in each group are approximately normally distributed. 3) Homoscedasticity (homogeneity of variances): The variance of the dependent variable is equal across all groups. Violations of these assumptions can affect the validity of the ANOVA results.



###21. What are the different types of ANOVA tests?
The different types of ANOVA tests include:

- One-Way ANOVA: Used to compare the means of three or more independent groups based on one independent variable.
- Two-Way ANOVA: Examines the effect of two independent categorical variables on a single dependent variable, and also investigates their interaction effect.
- MANOVA (Multivariate Analysis of Variance): Used when there are two or more dependent variables.
- Repeated Measures ANOVA: Used when the same subjects are measured multiple times under different conditions or over time.


###22. What is the F-test, and how does it relate to hypothesis testing?
The F-test is a statistical test based on the F-distribution, used to compare the variances of two populations or, more commonly, to assess the equality of means across multiple groups in an ANOVA. In hypothesis testing, the F-test calculates an F-statistic, which is the ratio of two variances (mean squares). A significant F-statistic (large F-value and small P-value) indicates that the variation between group means is significantly larger than the variation within groups, leading to the rejection of the null hypothesis that all group means are equal.

###2.Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

def generate_discrete_uniform_and_plot_pmf(low, high, num_samples):
    samples = np.random.randint(low, high + 1, num_samples)
    counts = Counter(samples)
    values = sorted(counts.keys())
    probabilities = [counts[val] / num_samples for val in values]

    plt.figure(figsize=(8, 6))
    plt.stem(values, probabilities)
    plt.title(f'Probability Mass Function (PMF) of Discrete Uniform Distribution (low={low}, high={high})')
    plt.xlabel('Value')
    plt.ylabel('Probability')
    plt.xticks(values)
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    generate_discrete_uniform_and_plot_pmf(low=1, high=6, num_samples=10000) # Simulating a fair die
```

###3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

```python
def bernoulli_pdf(k, p):
    if not (0 <= p <= 1):
        raise ValueError("Probability 'p' must be between 0 and 1.")
    if k == 1:
        return p
    elif k == 0:
        return 1 - p
    else:
        return 0

if __name__ == "__main__":
    p_success = 0.7
    prob_success = bernoulli_pdf(1, p_success)
    prob_failure = bernoulli_pdf(0, p_success)

    print(f"Bernoulli PDF for k=1 (success) with p={p_success}: {prob_success}")
    print(f"Bernoulli PDF for k=0 (failure) with p={p_success}: {prob_failure}")
    print(f"Bernoulli PDF for k=2 with p={p_success}: {bernoulli_pdf(2, p_success)}")
```

###6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_discrete_uniform_cdf(low, high):
    values = np.arange(low, high + 1)
    num_outcomes = high - low + 1
    probability_per_outcome = 1 / num_outcomes

    # Calculate CDF values
    cdf_values = []
    cumulative_prob = 0
    for i in range(len(values)):
        cumulative_prob += probability_per_outcome
        cdf_values.append(cumulative_prob)

    plt.figure(figsize=(10, 6))
    plt.step(values, cdf_values, where='post', marker='o', linestyle='-', color='blue')
    plt.title(f'Cumulative Distribution Function (CDF) of Discrete Uniform Distribution (low={low}, high={high})')
    plt.xlabel('Value (x)')
    plt.ylabel('F(x)')
    plt.xticks(values)
    plt.yticks(np.arange(0, 1.1, 0.1))
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

if __name__ == "__main__":
    calculate_and_plot_discrete_uniform_cdf(low=1, high=6)
```

###9. Write a Python function to calculate Z-scores from a dataset and plot them


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def calculate_and_plot_z_scores(data):
    data = np.asarray(data)
    mean = np.mean(data)
    std_dev = np.std(data)

    if std_dev == 0:
        print("Standard deviation is zero, Z-scores cannot be calculated (all values are the same).")
        return np.zeros_like(data)

    z_scores = (data - mean) / std_dev

    plt.figure(figsize=(10, 6))
    plt.hist(z_scores, bins=20, density=True, edgecolor='black', alpha=0.7)
    plt.title('Histogram of Z-Scores')
    plt.xlabel('Z-Score')
    plt.ylabel('Density')
    plt.grid(axis='y', alpha=0.75)

    # Optional: Plot a standard normal PDF for comparison
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = stats.norm.pdf(x, 0, 1)
    plt.plot(x, p, 'k', linewidth=2, label='Standard Normal PDF')
    plt.legend()
    plt.show()

    return z_scores

if __name__ == "__main__":
    # Example dataset
    data_set = [10, 12, 12, 13, 15, 17, 18, 18, 19, 20, 22, 23, 25, 27, 28, 30]
    z_scores_result = calculate_and_plot_z_scores(data_set)
    print("Calculated Z-Scores:", z_scores_result)

    # Another example with normally distributed data
    normal_data = np.random.normal(loc=50, scale=10, size=1000)
    z_scores_normal = calculate_and_plot_z_scores(normal_data)
    print("\nCalculated Z-Scores (Normal Data sample):", z_scores_normal[:10]) # Print first 10 for brevity
```

###10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.


```python
import numpy as np
import matplotlib.pyplot as plt

def implement_clt_for_non_normal_distribution(distribution_type='exponential', param=1.0, sample_size=30, num_samples_of_means=1000):
    sample_means = []

    for _ in range(num_samples_of_means):
        if distribution_type == 'exponential':
            # Generate sample from an Exponential distribution (non-normal)
            sample = np.random.exponential(scale=param, size=sample_size)
        elif distribution_type == 'uniform':
            # Generate sample from a Uniform distribution (non-normal)
            sample = np.random.uniform(low=0, high=param, size=sample_size)
        else:
            raise ValueError("Unsupported distribution_type. Choose 'exponential' or 'uniform'.")

        sample_means.append(np.mean(sample))

    plt.figure(figsize=(12, 6))

    # Plot histogram of the original non-normal distribution (a single large sample)
    if distribution_type == 'exponential':
        original_data = np.random.exponential(scale=param, size=100000)
        plt.subplot(1, 2, 1)
        plt.hist(original_data, bins=50, density=True, alpha=0.7, color='skyblue', edgecolor='black')
        plt.title(f'Original {distribution_type.capitalize()} Distribution (λ={param})')
        plt.xlabel('Value')
        plt.ylabel('Density')
    elif distribution_type == 'uniform':
        original_data = np.random.uniform(low=0, high=param, size=100000)
        plt.subplot(1, 2, 1)
        plt.hist(original_data, bins=50, density=True, alpha=0.7, color='lightgreen', edgecolor='black')
        plt.title(f'Original {distribution_type.capitalize()} Distribution (low=0, high={param})')
        plt.xlabel('Value')
        plt.ylabel('Density')


    # Plot histogram of the sample means
    plt.subplot(1, 2, 2)
    plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='lightcoral', edgecolor='black')
    plt.title(f'Distribution of Sample Means (n={sample_size})')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')

    plt.tight_layout()
    plt.show()

    print(f"Mean of sample means: {np.mean(sample_means):.4f}")
    print(f"Standard deviation of sample means (Standard Error): {np.std(sample_means):.4f}")

if __name__ == "__main__":
    print("Demonstrating CLT with Exponential Distribution:")
    implement_clt_for_non_normal_distribution(distribution_type='exponential', param=3.0, sample_size=30, num_samples_of_means=5000)

    print("\nDemonstrating CLT with Uniform Distribution:")
    implement_clt_for_non_normal_distribution(distribution_type='uniform', param=10.0, sample_size=40, num_samples_of_means=5000)
```

###11.Simulate multiple samples from a normal distribution and verify the Central Limit Theorem

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def verify_clt_normal_distribution(population_mean, population_std, sample_size, num_samples):
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.normal(population_mean, population_std, sample_size)
        sample_means.append(np.mean(sample))

    plt.figure(figsize=(12, 6))

    # Plot histogram of the original normal distribution
    plt.subplot(1, 2, 1)
    original_data = np.random.normal(population_mean, population_std, 100000)
    plt.hist(original_data, bins=50, density=True, alpha=0.7, color='skyblue', edgecolor='black')
    plt.title(f'Original Normal Distribution (μ={population_mean}, σ={population_std})')
    plt.xlabel('Value')
    plt.ylabel('Density')
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p_original = norm.pdf(x, population_mean, population_std)
    plt.plot(x, p_original, 'r--', linewidth=2, label='Population PDF')
    plt.legend()


    # Plot histogram of the sample means
    plt.subplot(1, 2, 2)
    plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='lightcoral', edgecolor='black')
    plt.title(f'Distribution of Sample Means (n={sample_size})')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')

    # Overlay theoretical normal distribution for sample means
    mean_of_sample_means = np.mean(sample_means)
    std_error = population_std / np.sqrt(sample_size) # Theoretical standard error
    xmin_means, xmax_means = plt.xlim()
    x_means = np.linspace(xmin_means, xmax_means, 100)
    p_means = norm.pdf(x_means, population_mean, std_error) # Use population_mean as theoretical mean
    plt.plot(x_means, p_means, 'g--', linewidth=2, label='Theoretical Normal PDF')
    plt.legend()

    plt.tight_layout()
    plt.show()

    print(f"Population Mean: {population_mean}")
    print(f"Population Standard Deviation: {population_std}")
    print(f"Sample Size (n): {sample_size}")
    print(f"Number of Samples of Means: {num_samples}")
    print("-" * 30)
    print(f"Observed Mean of Sample Means: {mean_of_sample_means:.4f}")
    print(f"Theoretical Mean of Sample Means (Population Mean): {population_mean:.4f}")
    print(f"Observed Standard Deviation of Sample Means: {np.std(sample_means):.4f}")
    print(f"Theoretical Standard Error (σ/√n): {std_error:.4f}")

if __name__ == "__main__":
    verify_clt_normal_distribution(population_mean=50, population_std=10, sample_size=30, num_samples=5000)
```

###13. Generate random variables and calculate their corresponding probabilities using the binomial distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def simulate_and_calculate_binomial_probabilities(n_trials, p_success, num_simulations):
    # 1. Simulate random variables
    simulated_outcomes = np.random.binomial(n_trials, p_success, num_simulations)

    # 2. Calculate theoretical probabilities for all possible outcomes (k)
    possible_outcomes = np.arange(0, n_trials + 1)
    theoretical_probabilities = binom.pmf(possible_outcomes, n_trials, p_success)

    print(f"Simulated {num_simulations} outcomes from Binomial(n={n_trials}, p={p_success}):")
    print(f"First 10 simulated outcomes: {simulated_outcomes[:10]}")
    print("\nTheoretical Probabilities (PMF) for each number of successes:")
    for k, prob in zip(possible_outcomes, theoretical_probabilities):
        print(f"P(X={k}) = {prob:.4f}")

    # Visualize the simulated vs. theoretical PMF
    plt.figure(figsize=(10, 6))
    plt.hist(simulated_outcomes, bins=np.arange(-0.5, n_trials + 1.5, 1),
             density=True, edgecolor='black', alpha=0.6, label='Simulated Frequencies')
    plt.stem(possible_outcomes, theoretical_probabilities, linefmt='r--', markerfmt='ro',
             basefmt=' ', label='Theoretical PMF')

    plt.title(f'Binomial Distribution Simulation vs. PMF (n={n_trials}, p={p_success})')
    plt.xlabel('Number of Successes (k)')
    plt.ylabel('Probability / Relative Frequency')
    plt.xticks(possible_outcomes)
    plt.grid(axis='y', alpha=0.75)
    plt.legend()
    plt.show()

if __name__ == "__main__":
    simulate_and_calculate_binomial_probabilities(n_trials=15, p_success=0.4, num_simulations=20000)
```

###14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_and_compare_z_score(data_point, dataset):
    dataset = np.asarray(dataset)
    mean = np.mean(dataset)
    std_dev = np.std(dataset)

    if std_dev == 0:
        print("Standard deviation of the dataset is zero. Cannot calculate Z-score meaningfully.")
        return None

    z_score = (data_point - mean) / std_dev

    print(f"Dataset Mean: {mean:.2f}")
    print(f"Dataset Standard Deviation: {std_dev:.2f}")
    print(f"Data Point: {data_point}")
    print(f"Calculated Z-Score: {z_score:.4f}")

    # Plotting the standard normal distribution
    x = np.linspace(-4, 4, 1000)
    pdf = norm.pdf(x, loc=0, scale=1)

    plt.figure(figsize=(10, 6))
    plt.plot(x, pdf, color='blue', linewidth=2, label='Standard Normal PDF')
    plt.fill_between(x, pdf, where=(x <= z_score), color='skyblue', alpha=0.3, label=f'P(Z <= {z_score:.2f})')

    plt.axvline(z_score, color='red', linestyle='--', label=f'Z-score = {z_score:.2f}')
    plt.title('Z-Score Comparison with Standard Normal Distribution')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.show()

    # You can also get the cumulative probability for this Z-score
    cumulative_prob = norm.cdf(z_score)
    print(f"Cumulative probability (P(Z <= {z_score:.4f})): {cumulative_prob:.4f}")

if __name__ == "__main__":
    # Example Dataset (e.g., test scores)
    my_dataset = np.array([65, 70, 72, 75, 78, 80, 82, 85, 88, 90, 92, 95])
    my_data_point = 85

    calculate_and_compare_z_score(my_data_point, my_dataset)

    # Another example
    my_data_point_2 = 60
    calculate_and_compare_z_score(my_data_point_2, my_dataset)
```

###15. Implement hypothesis testing using Z-statistics for a sample dataset

```python
import numpy as np
from scipy.stats import norm

def z_test_one_sample(sample_mean, population_mean, population_std_dev, sample_size, alpha=0.05, two_tailed=True):
    if sample_size <= 0 or population_std_dev <= 0:
        raise ValueError("Sample size and population standard deviation must be positive.")

    # Calculate the standard error of the mean
    standard_error = population_std_dev / np.sqrt(sample_size)

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / standard_error

    # Calculate the P-value
    if two_tailed:
        p_value = 2 * (1 - norm.cdf(abs(z_statistic)))
    else: # One-tailed (right-tailed)
        p_value = 1 - norm.cdf(z_statistic)
        # For left-tailed test, use norm.cdf(z_statistic) and adjust alternative hypothesis wording.

    print(f"--- Z-Test Results ---")
    print(f"Sample Mean: {sample_mean}")
    print(f"Hypothesized Population Mean (μ0): {population_mean}")
    print(f"Known Population Standard Deviation (σ): {population_std_dev}")
    print(f"Sample Size (n): {sample_size}")
    print(f"Calculated Z-Statistic: {z_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Significance Level (α): {alpha}")
    print(f"Test Type: {'Two-tailed' if two_tailed else 'One-tailed'}")

    # Make a decision
    if p_value <= alpha:
        decision = "Reject the null hypothesis."
        interpretation = "There is sufficient evidence to conclude that the sample mean is statistically different from the hypothesized population mean."
    else:
        decision = "Fail to reject the null hypothesis."
        interpretation = "There is not sufficient evidence to conclude that the sample mean is statistically different from the hypothesized population mean."

    print(f"Decision: {decision}")
    print(f"Interpretation: {interpretation}")

    return z_statistic, p_value, decision

if __name__ == "__main__":
    # Example 1: Two-tailed test
    # A factory produces bolts with a known std dev of 0.1mm.
    # A sample of 100 bolts has a mean length of 10.05mm.
    # Is this significantly different from the target length of 10mm?
    print("Example 1: Two-tailed Z-test (Is mean bolt length different from 10mm?)")
    z_test_one_sample(sample_mean=10.05, population_mean=10.0, population_std_dev=0.1, sample_size=100, alpha=0.05, two_tailed=True)
    print("\n" + "="*50 + "\n")

    # Example 2: One-tailed test (right-tailed)
    # A new teaching method is introduced. The historical average test score is 75 with a std dev of 10.
    # A sample of 50 students using the new method has an average score of 78.
    # Has the new method *increased* test scores?
    print("Example 2: One-tailed Z-test (Has new method *increased* test scores?)")
    z_test_one_sample(sample_mean=78, population_mean=75, population_std_dev=10, sample_size=50, alpha=0.05, two_tailed=False)
```

###16. Create a confidence interval for a dataset using Python and interpret the result


```python
import numpy as np
from scipy.stats import norm

def create_confidence_interval_z(sample_mean, population_std_dev, sample_size, confidence_level=0.95):
    if not (0 < confidence_level < 1):
        raise ValueError("Confidence level must be between 0 and 1.")
    if sample_size <= 0 or population_std_dev <= 0:
        raise ValueError("Sample size and population standard deviation must be positive.")

    # Calculate the critical Z-value
    alpha = 1 - confidence_level
    critical_z = norm.ppf(1 - alpha / 2) # For two-tailed interval

    # Calculate the standard error of the mean
    standard_error = population_std_dev / np.sqrt(sample_size)

    # Calculate the margin of error
    margin_of_error = critical_z * standard_error

    # Calculate the confidence interval bounds
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    interpretation = (
        f"We are {confidence_level*100:.0f}% confident that the true population mean "
        f"lies within the interval [{lower_bound:.3f}, {upper_bound:.3f}]."
    )

    print(f"--- Confidence Interval (Z-based) ---")
    print(f"Sample Mean: {sample_mean:.3f}")
    print(f"Known Population Standard Deviation: {population_std_dev:.3f}")
    print(f"Sample Size: {sample_size}")
    print(f"Confidence Level: {confidence_level*100:.0f}%")
    print(f"Critical Z-value: {critical_z:.4f}")
    print(f"Standard Error of the Mean: {standard_error:.4f}")
    print(f"Margin of Error: {margin_of_error:.4f}")
    print(f"Confidence Interval: [{lower_bound:.3f}, {upper_bound:.3f}]")
    print(f"Interpretation: {interpretation}")

    return lower_bound, upper_bound, margin_of_error, interpretation

if __name__ == "__main__":
    # Example: A quality control manager wants to estimate the average weight of cereal boxes.
    # From past experience, the population standard deviation is known to be 5 grams.
    # A sample of 50 boxes has a mean weight of 360 grams.
    # Calculate a 95% confidence interval for the true average weight.

    sample_mean_weight = 360
    pop_std_dev_weight = 5
    sample_size_boxes = 50
    conf_level = 0.95

    create_confidence_interval_z(
        sample_mean=sample_mean_weight,
        population_std_dev=pop_std_dev_weight,
        sample_size=sample_size_boxes,
        confidence_level=conf_level
    )

    print("\n" + "="*50 + "\n")

    # Example 2: Using a larger sample size for demonstration
    create_confidence_interval_z(
        sample_mean=150,
        population_std_dev=15,
        sample_size=200,
        confidence_level=0.99
    )
```

###17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean


```python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt

def generate_and_interpret_confidence_interval_t(population_mean, population_std, sample_size, confidence_level=0.95):
    # 1. Generate data from a normal distribution
    sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

    # 2. Calculate sample statistics
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1) # Use ddof=1 for sample standard deviation

    # 3. Calculate the critical t-value
    alpha = 1 - confidence_level
    degrees_of_freedom = sample_size - 1
    critical_t = t.ppf(1 - alpha / 2, degrees_of_freedom)

    # 4. Calculate the standard error of the mean
    standard_error = sample_std / np.sqrt(sample_size)

    # 5. Calculate the margin of error
    margin_of_error = critical_t * standard_error

    # 6. Calculate the confidence interval bounds
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    interpretation = (
        f"We are {confidence_level*100:.0f}% confident that the true population mean "
        f"lies within the interval [{lower_bound:.3f}, {upper_bound:.3f}]."
    )

    print(f"--- Confidence Interval (T-based) for Generated Data ---")
    print(f"Actual Population Mean (for reference): {population_mean:.3f}")
    print(f"Actual Population Std Dev (for reference): {population_std:.3f}")
    print("-" * 40)
    print(f"Generated Sample Size (n): {sample_size}")
    print(f"Generated Sample Mean: {sample_mean:.3f}")
    print(f"Generated Sample Standard Deviation: {sample_std:.3f}")
    print(f"Confidence Level: {confidence_level*100:.0f}%")
    print(f"Degrees of Freedom: {degrees_of_freedom}")
    print(f"Critical t-value: {critical_t:.4f}")
    print(f"Standard Error of the Mean: {standard_error:.4f}")
    print(f"Margin of Error: {margin_of_error:.4f}")
    print(f"Confidence Interval: [{lower_bound:.3f}, {upper_bound:.3f}]")
    print(f"Interpretation: {interpretation}")

    # Visualize the confidence interval
    plt.figure(figsize=(10, 5))
    plt.errorbar(x=sample_mean, y=0, xerr=margin_of_error, fmt='o', color='blue', capsize=5, label='Confidence Interval')
    plt.axvline(x=population_mean, color='red', linestyle='--', label='True Population Mean')
    plt.title(f'{confidence_level*100:.0f}% Confidence Interval for Mean (n={sample_size})')
    plt.xlabel('Mean Value')
    plt.yticks([]) # Hide y-axis ticks
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.legend()
    plt.show()

if __name__ == "__main__":
    generate_and_interpret_confidence_interval_t(
        population_mean=100, population_std=15, sample_size=30, confidence_level=0.95
    )
    print("\n" + "="*50 + "\n")
    generate_and_interpret_confidence_interval_t(
        population_mean=25, population_std=5, sample_size=15, confidence_level=0.90
    )
```

###19. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def calculate_and_interpret_poisson_cdf(lambda_param, max_k):
    if lambda_param <= 0:
        raise ValueError("Lambda (λ) must be positive for Poisson distribution.")
    if max_k < 0:
        raise ValueError("max_k must be non-negative.")

    # Create a range of k values (number of events)
    k_values = np.arange(0, max_k + 1)

    # Calculate the CDF for each k
    cdf_values = poisson.cdf(k_values, mu=lambda_param)

    print(f"--- Poisson CDF (λ={lambda_param}) ---")
    for k, cdf_val in zip(k_values, cdf_values):
        print(f"P(X <= {k}) = {cdf_val:.4f}")

    # Plotting the CDF
    plt.figure(figsize=(10, 6))
    plt.step(k_values, cdf_values, where='post', marker='o', linestyle='-', color='purple')
    plt.title(f'Cumulative Distribution Function (CDF) of Poisson Distribution (λ={lambda_param})')
    plt.xlabel('Number of Events (k)')
    plt.ylabel('P(X <= k)')
    plt.xticks(k_values)
    plt.yticks(np.arange(0, 1.1, 0.1))
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

    # Interpretation example
    interpret_k = int(max_k / 2) # Pick a point for interpretation
    if interpret_k >= 0:
        prob_less_than_eq_interpret_k = poisson.cdf(interpret_k, mu=lambda_param)
        print(f"\nInterpretation:")
        print(f"The probability of observing up to {interpret_k} events is {prob_less_than_eq_interpret_k:.4f}.")
        print(f"This means in a long run, {prob_less_than_eq_interpret_k*100:.2f}% of the time, you would expect to see {interpret_k} or fewer events.")


if __name__ == "__main__":
    calculate_and_interpret_poisson_cdf(lambda_param=2, max_k=8)
    print("\n" + "="*50 + "\n")
    calculate_and_interpret_poisson_cdf(lambda_param=5, max_k=12)
```

```python
def visualize_errors():
    x = np.linspace(-4, 4, 1000)
    h0 = norm.pdf(x, 0, 1)
    h1 = norm.pdf(x, 1.5, 1)

    plt.plot(x, h0, label='H0', color='blue')
    plt.plot(x, h1, label='H1', color='orange')
    plt.fill_between(x, h0, where=(x > 1.645), color='red', alpha=0.3, label='Type I Error')
    plt.fill_between(x, h1, where=(x <= 1.645), color='green', alpha=0.3, label='Type II Error')
    plt.axvline(1.645, color='black', linestyle='--')
    plt.legend()
    plt.title("Type I and II Errors")
    plt.xlabel("Test Statistic")
    plt.ylabel("Density")
    plt.grid(True)
    plt.show()

visualize_errors()
```

###9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance

```python
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    stderr = np.std(data, ddof=1) / np.sqrt(n)
    h = t.ppf((1 + confidence) / 2, n - 1) * stderr
    return mean - h, mean + h

data = np.random.normal(70, 8, 40)
ci = confidence_interval(data)
print(f"95% Confidence Interval: {ci}")
```

###10.Write a Python program to calculate the margin of error for a given confidence level using sample data

```python
def margin_of_error(data, confidence=0.95):
    n = len(data)
    stderr = np.std(data, ddof=1) / np.sqrt(n)
    moe = t.ppf((1 + confidence) / 2, n - 1) * stderr
    return moe

print("Margin of Error:", margin_of_error(data))
```

###11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

```python
def bayes_theorem(p_a, p_b_given_a, p_b_given_not_a):
    p_not_a = 1 - p_a
    p_b = p_b_given_a * p_a + p_b_given_not_a * p_not_a
    p_a_given_b = (p_b_given_a * p_a) / p_b
    return p_a_given_b

# Example: Disease test
posterior = bayes_theorem(0.01, 0.95, 0.05)
print(f"Posterior Probability: {posterior:.4f}")
```
